```python
# Combining all 18 datasets
import pandas as pd
import numpy as np

def load_and_process(suffix):
    # Load datasets
    downlink = pd.read_csv(f"ch_d_{suffix}.csv", header=None)
    uplink = pd.read_csv(f"ch_u_{suffix}.csv", header=None)
    crosslink = pd.read_csv(f"ch_ud_{suffix}.csv", header=None)
    transmitpower = pd.read_csv(f"P_BS_{suffix}.csv", header=None)
    userpower = pd.read_csv(f"P_{suffix}.csv", header=None)
    user = pd.read_csv(f"user_d_{suffix}.csv", header=None)

    # Align dimensions
    crosslink = crosslink.iloc[:54, :]  # trim to match user count

    # Convert crosslink to dB
    crosslink_db = 10 * np.log10(crosslink.replace(0, np.nan))

    # --- Aggregate features ---
    downlink_features = pd.DataFrame({
        'avg_downlink': downlink.mean(axis=1),
        'std_downlink': downlink.std(axis=1),
        'min_downlink': downlink.min(axis=1),
        'max_downlink': downlink.max(axis=1)
    })

    uplink_features = pd.DataFrame({
        'avg_uplink': uplink.mean(axis=1),
        'std_uplink': uplink.std(axis=1),
        'min_uplink': uplink.min(axis=1),
        'max_uplink': uplink.max(axis=1)
    })

    crosslink_features = pd.DataFrame({
        'avg_crosslink': crosslink_db.mean(axis=1),
        'std_crosslink': crosslink_db.std(axis=1),
        'min_crosslink': crosslink_db.min(axis=1),
        'max_crosslink': crosslink_db.max(axis=1)
    })

    userpower_features = pd.DataFrame({
        'avg_userpower': userpower.mean(axis=1),
        'std_userpower': userpower.std(axis=1),
        'min_userpower': userpower.min(axis=1),
        'max_userpower': userpower.max(axis=1)
    })

    transmitpower_features = pd.DataFrame({
        'avg_transmitpower': transmitpower.mean(axis=1),
        'std_transmitpower': transmitpower.std(axis=1),
        'min_transmitpower': transmitpower.min(axis=1),
        'max_transmitpower': transmitpower.max(axis=1)
    })

    # --- Estimate interference (targets) ---
    interference_downlink = (crosslink.values * userpower.values).mean(axis=1)
    interference_uplink = (downlink.values * transmitpower.values).mean(axis=1)

    # --- Combine all ---
    df = pd.concat([
        downlink_features, uplink_features, crosslink_features,
        transmitpower_features, userpower_features
    ], axis=1)

    df['avg_downlink_interference'] = interference_downlink
    df['avg_uplink_interference'] = interference_uplink
    df['scenario'] = suffix  # label scenario (A3, A4, A5)

    return df
```

```python
scenarios = ['A3', 'A4', 'A5']
combined = [load_and_process(d) for d in scenarios]
dataset_all = pd.concat(combined, ignore_index=True)
```

# Feature Engineering

```python
dataset_all['scenario_code'] = dataset_all['scenario'].map({'A3': 3, 'A4': 4, 'A5': 5})
```

```python
dataset_all.head(100)
```

```python
dataset_all['donw_to_up'] = dataset_all['avg_downlink'] / dataset_all['avg_uplink'] # How strong downlink is relative to uplink
dataset_all['BS_to_A_power'] = dataset_all['avg_transmitpower'] / dataset_all['avg_userpower'] #Base station power dominance
dataset_all['relative_inter'] = dataset_all['avg_crosslink'] / dataset_all['avg_downlink'] # Relative interference strength
dataset_all['ch_ud_to_ch_u'] = dataset_all['avg_crosslink'] / dataset_all['avg_uplink'] # Crosslink interference seen from uplink side
```

```python
dataset_all['log_DL_interference'] = np.log1p(dataset_all['avg_downlink_interference'])
dataset_all['log_UL_interference'] = np.log1p(dataset_all['avg_uplink_interference'])
```

```python
dataset_all['DL_efficiency'] = dataset_all['avg_downlink_interference'] / dataset_all['avg_userpower']
dataset_all['UL_efficiency'] = dataset_all['avg_uplink_interference'] / dataset_all['avg_transmitpower']
```

```python
# 1e-6 avoids division by zero
dataset_all['approx_DL_SINR'] = dataset_all['avg_downlink'] / (1e-6 + dataset_all['avg_downlink_interference'])
dataset_all['approx_UL_SINR'] = dataset_all['avg_uplink'] / (1e-6 + dataset_all['avg_uplink_interference'])
```

# Training Models for uplink

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import seaborn as sns
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
```

```python
# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "ElasticNet": ElasticNet(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42, objective='reg:squarederror'),
    # "MLP Regressor": MLPRegressor(random_state=42, max_iter=2000)
}

# Function to evaluate model
def evaluate_model_uplink(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    r2 = r2_score(y_test, preds)
    mape = np.mean(np.abs((y_test - preds) / y_test)) * 100

    return mae, rmse, r2, mape, preds

# Evaluate all models
results = []
predictions = {}

for name, model in models.items():
    mae, rmse, r2, mape, preds = evaluate_model_uplink(model, X_train_ul, X_test_ul, y_train_ul, y_test_ul)
    results.append([name, mae, rmse, r2, mape])
    predictions[name] = preds
    print(f"{name}: R2={r2:.3f}, RMSE={rmse:.5f}")

# Create results dataframe
results_df_uplink = pd.DataFrame(results, columns=["Model", "MAE", "RMSE", "RÂ²", "MAPE"])
results_df_uplink.sort_values(by="RÂ²", ascending=False, inplace=True)
results_df_uplink
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="RÂ²", y="Model", data=results_df_uplink, palette="viridis")
plt.title("Model RÂ² Comparison - Uplink Interference")
plt.xlabel("RÂ² Score (Higher is Better)")
plt.ylabel("")
plt.show()
```

#hyperparameter tuning for uplink

```python
# Define Ridge model
ridge = Ridge()

# Define parameter distributions (continuous and categorical)
param_dist = {
    'alpha': uniform(0.0001, 100),       # random values between 0.0001 and ~100
    'solver': ['auto', 'svd', 'lsqr', 'saga'],
    'fit_intercept': [True, False]
}

# Randomized search setup
ridge_random = RandomizedSearchCV(
    estimator=ridge,
    param_distributions=param_dist,
    n_iter=30,               # number of random combinations to try
    scoring='r2',
    cv=5,
    n_jobs=-1,
    random_state=42,
    verbose=2
)

# Fit model
ridge_random.fit(X_train_ul, y_train_ul)

# Best model
best_ridge = ridge_random.best_estimator_

# Evaluate on test set
y_pred_uplink = best_ridge.predict(X_test_ul)

mae = mean_absolute_error(y_test_ul, y_pred_uplink)
rmse = np.sqrt(mean_squared_error(y_test_ul, y_pred_uplink))
r2 = r2_score(y_test_ul, y_pred_uplink)

print("âœ… Best Ridge Model Parameters:", ridge_random.best_params_)
print(f"MAE: {mae:.8f}, RMSE: {rmse:.8f}, RÂ²: {r2:.8f}")
```
